```python

from animal_shelter import AnimalShelter

s = AnimalShelter("aacuser", "pass", "nv-desktop-services.apporto.com", 30611)

assert s.create({"animal_id":"X1","animal_type":"Cat","breed":"Domestic Shorthair"}) is True
assert isinstance(s.read({"animal_type":"Cat"}), list)
```

```python
print("Create returned:", s.create({
    "animal_id":   "X1",
    "animal_type": "Cat",
    "breed":       "Domestic Shorthair"
}))
print("Read returned type:", type(s.read({"animal_type":"Cat"})))
```

```python

num_updated = s.update(
    {"animal_id": "X1"},
    {"outcome_type": "Adoption"}
)

assert num_updated > 0

print("Update returned:", num_updated)
print("Post-update document:", s.read({"animal_id": "X1"}))
```

```python

num_deleted = s.delete({"animal_id": "X1"})


assert num_deleted > 0

print("Delete returned:", num_deleted)
print("Remaining docs:", s.read({"animal_id": "X1"}))
```
